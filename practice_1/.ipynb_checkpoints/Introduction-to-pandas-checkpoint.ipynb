```python
import pandas as pd
```

There are Series and DataFrame in Pandas. `Series` take python list and `DataFrame` take python dictionary.

```python
series = pd.Series(["BMW", "Tesla", "Honda"])
series
```

Series is one dimensional

```python
colors = pd.Series(["Blue", "Red", "White"])
colors
```

DataFrame is two dimensional

```python
car_data = pd.DataFrame({
    "Car make": series,
    "Color": colors
})

car_data
```
